```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
# import numpy as np
# np.random.seed(3)
# x = 3 + np.random.normal(0, 2, 50)
# y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
```

```python
x = 3+np.random.normal(0,2,50)
y = 3+np.random.normal(0,2,len(x))
```

```python
y
```

```python
plt.scatter(x,y,c = 'blue',label = 'Data Points')
plt.title("Scatter Plot of x vs y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# Q.3 Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2= np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(12,8))
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1,label = 'Line 1')
plt.title("Line 1")
```

```python
plt.subplot(2,2,2)
plt.plot(x,y2,label = 'Line 2')
plt.title("Line 2")
```

```python
plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3')
plt.title('Line 3')
```

```python
plt.subplot(2,2,4)
plt.plot(x,y4,label = 'Line 4')
plt.title("Line 4")
```

```python
plt.legend()
```

```python
plt.tight_layout()
```

```python
plt.show()
```

```python
# What is a box plot? Why is it used? Using the following data plot a box plot.
# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)
```

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)
```

```python
plt.figure(figsize=(8,6))
plt.boxplot([box1,box2],labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Box 1 and Box 2')
plt.ylabel('Values')
```
