```python
# Q5. Explain Primary Key and Foreign Key.
# Ans : -
#  1.Primary Key:

# A Primary Key is a column or set of columns that uniquely identifies each record (row) in a table.
# It must contain unique values for each row and cannot contain NULL values.
# Every table in a relational database should have a Primary Key, as it serves as a unique identifier for the records.
# Typically, Primary Keys are implemented using unique indexes.
# Examples of Primary Key columns include employee ID, student ID, order ID, etc.
# 2. Foreign Key:

# A Foreign Key is a column or set of columns in a table that refers to the Primary Key of another table.
# It establishes a relationship between two tables by referencing the Primary Key column(s) of one table in another table.
# Foreign Keys help enforce referential integrity, ensuring that data in related tables remain consistent.
# When a Foreign Key is defined in a table, the values in that column must match the values of the Primary Key column(s) in the referenced table, or be NULL (if the Foreign Key column allows NULL values).
# Foreign Keys can be one-to-one, one-to-many, or many-to-many relationships between tables.
# Examples of Foreign Key columns include employee ID referencing a department table, customer ID referencing an order table, etc.
# Here's an example illustrating the use of Primary Key and Foreign Key:
# Consider two tables: students and courses.
# sql
# Copy code
# CREATE TABLE students (
#     student_id INT PRIMARY KEY,
#     student_name VARCHAR(50),
#     age INT
# );

# CREATE TABLE courses (
#     course_id INT PRIMARY KEY,
#     course_name VARCHAR(50),
#     instructor VARCHAR(50),
#     student_id INT,
#     FOREIGN KEY (student_id) REFERENCES students(student_id)
# );
```

```python
# 7.Give the order of execution of SQL clauses in an SQL query.
# In SQL, the order of execution of clauses in an SQL query generally follows this sequence:

# 1.FROM: The FROM clause specifies the tables from which data will be retrieved or manipulated. It's the first clause processed in a query.

# 2.WHERE: The WHERE clause filters the rows returned by the FROM clause based on specified conditions. It's applied after the FROM clause and before any other clauses (except FROM itself).

# 3.GROUP BY: The GROUP BY clause is used to group rows that have the same values into summary rows. It's applied after the WHERE clause and before the SELECT clause.

# 4.HAVING: The HAVING clause filters grouped rows returned by the GROUP BY clause based on specified conditions. It's applied after the GROUP BY clause and before the ORDER BY clause.

# 5.SELECT: The SELECT clause specifies the columns that will be returned in the result set. It's applied after the previous clauses have been processed.

# 6.ORDER BY: The ORDER BY clause sorts the rows returned by the query based on specified columns or expressions. It's applied after the SELECT clause.

# 7.LIMIT/OFFSET: The LIMIT and OFFSET clauses are used to restrict the number of rows returned by the query and to skip a specific number of rows, respectively. They're typically applied after the ORDER BY clause.
```
